- 문제 1
- 다음은 A그룹과 B그룹 인원의 키 데이터이다.
- 두 그룹의 키 평균이 다르다고 할 수 있는지
- 가설검정을 실시하고자 한다. 아래 물음에 답하시오(유의수준 1%)
- A : A그룹 인원의 키 평균, B : B그룹 인원의 키 평균
- H0(귀무가설) : A = B
- H1(대립가설) : A != B

- 문제 1-1
- 독립표본 t검정을 실시하고 검정통계량, p-value값을 구하시오.
- 정답은 반올림하여 소수점 셋째자리까지 구하시오.
- (단, 두 그룹의 데이터는 정규성은 만족하지만, 등분산성은 만족하지 못한다고 가정한다.)
- 
- 문제 1-2
- 귀무가설 기각여부를 결정하시오(답은 채택 또는 기각으로 작성하시오)

```python
 ###############  실기환경 복사 영역  ###############
import pandas as pd
 # 데이터 생성
data = {
'Height': [172, 175, 173, 174, 177, 170, 169, 178, 171, 176,  # Group A
168, 165, 171, 169, 170, 174, 171, 171, 168, 175], # Group B
'Group': ['A'] * 10 + ['B'] * 10
 }
df = pd.DataFrame(data)
 ###############  실기환경 복사 영역  ###############
df.head(3)
A = df[df['Group']=='A']['Height']
B = df[df['Group']=='B']['Height']
# print(A)
# print(B)
# 1. 가설설정
# H0 : A = B
# H1 : A != B
# 2. 유의수준 확인 : 유의수준 1%로 확인
# 3. 정규성 검정
import scipy.stats as stats
from scipy.stats import shapiro
# H0 : 정규분포를 따른다
# H1 : 정규분포를 따르지 않는다
statisticA, pvalueA = shapiro(A)
statisticB, pvalueB = shapiro(B)
# print(round(statisticA, 3), round(pvalueA, 3))
# print(round(statisticB, 3), round(pvalueB, 3))
# 정규분포를 따른다
# 4. 등분산 검정
from scipy.stats import bartlett
statistic, pvalue = stats.bartlett(A, B)
#print(statistic, pvalue)
# 등분산성 만족이지만 문제에서 만족하지 못한다고 가정
# 5. 검정실시 (정규성O, 등분산X)
statistic, pvalue = stats.ttest_ind(A, B, equal_var=False)
print(round(statistic, 3), round(pvalue, 3))
# 검정통계량 2.474, pvalue 0.024
# (정규성X)
statistic, pvalue = stats.ranksums(A, B)
print(round(statistic, 3), round(pvalue, 3))
# 6. 귀무가설 기각여부 결정
# p-value 가 0.01보다 크므로 귀무가설 채택
# 즉, A와 B가 같다고 할 수 있다.
# 답 : 채택
```

- 문제 2
- 다음은 A,B,C 그룹 인원의 영어 성적 데이터이다.
- 세 그룹의 성적 평균이 같다고 할 수 있는지 ANOVA 분석을 실시하시오. (유의수준 5%)
- A,B,C : 각 그룹 인원의 성적
- H0(귀무가설) : A(평균) = B(평균) = C(평균)
- H1(대립가설) : not H0(적어도 하나는 같지 않다)
-
- 문제 2-1
- ANOVA 분석을 실시하고 검정통계량, p-value 값을 구하시오.
- 정답은 반올림하여 소수점 셋째자리까지 구하시오.
- (단, 각 그룹의 데이터는 정규성, 등분산성을 만족한다고 가정한다)
-
- 문제 2-2
- 귀무가설 기각여부를 결정하시오(답은 채택 또는 기각으로 작성하시오)  

```python
###############  실기환경 복사 영역  ###############
import pandas as pd
 # 데이터 생성
data = {
 'English_Score': [82, 84, 83, 85, 86, 87, 85, 84, 86, 88,  # Group A
 78, 77, 79, 76, 75, 77, 78, 79, 80, 76,  # Group B
 81, 83, 82, 84, 85, 83, 84, 82, 85, 86], # Group C
 'Group': ['A'] * 10 + ['B'] * 10 + ['C'] * 10
 }
df = pd.DataFrame(data)
 ###############  실기환경 복사 영역  ###############
#df.head(3)
#print(df)
A = df[df['Group']=='A']['English_Score']
B = df[df['Group']=='B']['English_Score']
C = df[df['Group']=='C']['English_Score']
# print(A)
# print(B)
# print(C)

# 1. 가설설정
# H0 : A = B = C
# H1 : not H0
# 2. 유의수준 확인 : 유의수준 5%로 확인
# 3. 정규성 검정
import scipy.stats as stats
from scipy.stats import shapiro
statisticA, pvalueA = shapiro(A)
statisticB, pvalueB = shapiro(B)
statisticC, pvalueC = shapiro(C)
# print(round(statisticA, 3), round(pvalueA, 3))
# print(round(statisticB, 3), round(pvalueB, 3))
# print(round(statisticC, 3), round(pvalueC, 3))
# 정규분포를 따른다
# 4. 등분산 검정
from scipy.stats import bartlett
statistic, pvalue = stats.bartlett(A,B,C)
#print(round(statistic, 3), round(pvalue, 3))
# 등분산성 만족
# 5. 검정실시 (정규성O, 등분산O)
statistic, pvalue = stats.f_oneway(A,B,C)
print(round(statistic, 3), round(pvalue, 3))
# 검정통계량 56.7, pvalue 0.0
# (정규성X)
statistic, pvalue = stats.kruskal(A,B,C)
print(round(statistic, 3), round(pvalue, 3))
# 6. 귀무가설 기각여부 결정
# p-value가 0.05보다 작으므로 귀무가설 기각
# 즉, 어느 하나라도 집단의 평균이 같지가 않다
# 답 : 기각
```

- 문제 3
- 어느 그룹에서 성별에 따라 선택한 스포츠가 관련성이 있는지 검정해보고자 한다.
- 두 변수(Sport, Gender)의 독립성 검정을 실시하시오. (유의수준 5%)
- Gender : Male, Female
- Sport : Soccer, Basketball, Swimming
- H0(귀무가설) : 두 변수는 서로 독립이다
- H1(대립가설) : 두 변수는 서로 독립이 아니다
- 
- 문제 3-1
- 카이제곱검정(독립성검정)을 실시하고 검정통계량, p-value 값을 구하시오.
- 정답은 반올림하여 소수점 셋째자리까지 구하시오.
-
- 문제 3-2
- 귀무가설 기각여부를 결정하시오.(답은 채택 또는 기각으로 작성하시오.) 

```python
###############  실기환경 복사 영역  ###############
import pandas as pd
import numpy as np
# 데이터 생성
data = {
    'Sport': ['Soccer'] * 35 + ['Basketball'] * 55 + ['Swimming'] * 60,
    'Gender': ['Male'] * 20 + ['Female'] * 15 + ['Male'] * 30 + ['Female'] * 25 + ['Male'] * 25 + ['Female'] * 35
}
df = pd.DataFrame(data)
###############  실기환경 복사 영역  ###############
#df.head(3)
#print(df.info())
# print(df['Sport'].value_counts())
# print(df['Gender'].value_counts())
#df['Gender'] = df['Gender'].map({'Female':1, 'Male':0})
A = df['Gender']
B = df['Sport']
table = pd.crosstab(index=A, columns=B)
print(table)
# 1. 가설설정
# H0 : 두 변수는 서로 영향을 주지 않는다 독립이다 p = 0
# H1 : 두 변수는 서로 영향을 준다 독립이 아니다 p != 0
# 2. 유의수준 확인 : 유의수준 5%로 확인
# 3. 검정실시
import scipy.stats as stats
from scipy.stats import chi2_contingency
statistic, pvalue, dof, expected = chi2_contingency([table.iloc[0], table.iloc[1]])
print(np.round(statistic, 3))
print(np.round(pvalue, 3))
print(dof)
print(np.round(expected, 3))
# 검정통계량 2.835 p-value 0.242
# 4. 귀무가설 기각여부 결정
# p-value가 0.05보다 크므로 귀무가설 채택
# 즉, 두 변수는 서로 영향을 주지 않는다 독립이다
# 답 : 채택
```
