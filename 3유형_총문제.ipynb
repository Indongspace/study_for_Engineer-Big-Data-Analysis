### 문제 1.
mtcars 데이터셋의 mpg열 데이터의 평균이 20과 같다고 할 수 있는지 검정하시오. (유의수준 5%)

```python
import pandas as pd
import numpy as np
df = pd.read_csv('mtcars.csv')
# print(df.head())

# 1. 가설설정
# H0 : mpg열 데이터의 평균은 20과 같다 (mu = 20)
# H1 : mpg열 데이터의 평균은 20과 같지 않다 (mu != 20)
# 2. 유의수준 확인 : 유의수준 5%로 확인
# 3. 정규성 검정
# H0 : 정규분포를 따른다
# H1 : 정규분포를 따르지 않는다
import scipy.stats as stats
from scipy.stats import shapiro
statistic, pvalue = stats.shapiro(df['mpg'])
# print(round(statistic, 4), round(pvalue, 4))
# 귀무가설 채택
# 4.1 검정 실시 (정규성O)
statistic, pvalue = stats.ttest_1samp(df['mpg'], popmean=20, alternative='two-sided')
print(round(statistic, 4), round(pvalue, 4))
# 4.2 (정규성X)
statistic2, pvalue2 = stats.wilcoxon(df['mpg']-20, alternative='two-sided')
print(round(statistic2, 4), round(pvalue2, 4))
# 5. 귀무가설 기각여부 결정
# p-value가 0.05보다 크므로 귀무가설 채택
# 즉, mpg열 데이터의 평균은 20과 같다고 할 수 있다.
# 답 : 채택
```

### 문제 4.
다음은 혈압약을 먹은 전,후의 혈압 데이터이다.
혈압약을 먹기 전, 후의 차이가 있는지 쌍체 t검정을 실시하시오. (유의수준 5%)
before : 혈압약을 먹기 전 혈압, after : 혈압약을 먹은 후의 혈압

```python
import pandas as pd
import numpy as np
#	데이터	만들기
df	=	pd.DataFrame(	{
	'before':	[120,	135,	122,	124,	135,	122,	145,	160,	155,	142,	144,	135,	167],
	'after'	:	[110,	132,	123,	119,	123,	115,	140,	162,	142,	138,	135,	142,	160]	})
#print(df.head(3))
# 1. 가설설정
# H0 : 혈압약을 먹기 전, 후의 차이가 없다 after - before = 0
# H1 : 혈압약을 먹기 전, 후의 차이가 있다 after - before != 0
# 2. 유의수준 확인 : 유의수준 5%
# 3. 정규성 확인
# H0 : 차이값이 정규분포를 따른다.
# H1 : 차이값이 정규분포를 따르지 않는다.
import scipy.stats as stats
from scipy.stats import shapiro
statistic, pvalue = stats.shapiro(df['after']-df['before'])
#print(round(statistic, 4), round(pvalue, 4))
# 정규분포를 따른다
# 4-1. 검정실시 (정규성O)
statistic, pvalue = stats.ttest_rel(df['after'], df['before'], alternative='two-sided')
print(round(statistic, 4), round(pvalue, 4))
# 4-2. (정규성X)
statistic2, pvalue2 = stats.wilcoxon(df['after']-df['before'], alternative='two-sided')
print(round(statistic2, 4), round(pvalue2, 4))
# 5. 귀무가설 기각여부 결정
# p-value 가 0.05보다 작으므로 귀무가설 기각
# 즉, 혈압약을 먹기 전, 후의 차이가 있다
# 답 : 기각
```

```python
 #	데이터	만들기
df	=	pd.DataFrame(	{
	'before':	[120,	135,	122,	124,	135,	122,	145,	160,	155,	142,	144,	135,	167],
	'after'	:	[110,	132,	123,	119,	123,	115,	140,	162,	142,	138,	135,	142,	160]	})
#print(df.head(3))
# 1. 가설설정
# H0 : 혈압약을 먹은후 혈압이 같거나 증가했다. after - before >= 0
# H1 : 혈압약을 먹은후 혈압이 감소했다. after - before < 0
# 2. 유의수준 확인 : 유의수준 5%로 확인
# 3. 정규성 검정
# H0 : 차이값이 정규분포를 따른다
# H1 : 차이값이 정규분포를 따르지 않는다
import scipy.stats as stats
from scipy.stats import shapiro
statistic, pvalue = stats.shapiro(df['after']-df['before'])
#print(round(statistic, 4), round(pvalue, 4))
# 정규분포를 따른다.
# 4-1. 검정실시 (정규성O)
statistic, pvalue = stats.ttest_rel(df['after'], df['before'], alternative='less')
print(round(statistic, 4), round(pvalue, 4))
# 4-2. (정규성X)
statistic2, pvalue2 = stats.wilcoxon(df['after']-df['before'], alternative='less')
print(round(statistic2, 4), round(pvalue2, 4))
# 5. 귀무가설 기각여부 결정
# p-value가 0.05보다 작으므로 귀무가설 기각
# 즉, 혈압약을 먹은 후 혈압이 감소했다.
# 답 : 기각
```

### 문제 6.
다음은 A그룹과 B그룹 인원의 혈압 데이터이다.
두 그룹의 혈압평균이 다르다고 할 수 있는지 독립표본 t검정을 실시하시오. (유의수준 5%)

```python
#	데이터	만들기
df = pd.DataFrame({
				'A': [120,	135,	122,	124,	135,	122,	145,	160,	155,	142,	144,	135,	167],
				'B'	: [110,	132,	123,	119,	123,	115,	140,	162,	142,	138,	135,	142,	160]})
#print(df.head(3))
# 1. 가설설정
# H0 : 두 그룹의 혈압평균이 같다 A = B
# H1 : 두 그룹의 혈압평균이 다르다 A != B
# 2. 유의수준 확인 : 유의수준 5%로 확인
# 3. 정규성 검정
# H0 : 두 그룹 모두 정규분포를 따른다
# H1 : not H0
import scipy.stats as stats
from scipy.stats import shapiro
statisticA, pvalueA = stats.shapiro(df['A'])
statisticB, pvalueB = stats.shapiro(df['B'])
print(round(statisticA, 4), round(pvalueA, 4))
print(round(statisticB, 4), round(pvalueB, 4))
# 두 그룹 모두 정규분포를 따른다
# 4. 등분산 검정
# H0 : 등분산한다
# H1 : 등분산하지 않는다
statistic, pvalue = stats.bartlett(df['A'], df['B'])
print(round(statistic, 4), round(pvalue, 4))
# 등분산한다
# 5-1. 검정실시 (정규성O, 등분산O)
statistic, pvalue = stats.ttest_ind(df['A'], df['B'], equal_var=True, alternative='two-sided')
print(round(statistic, 4), round(pvalue, 4))
# 5-2. (정규성X) 순위합검정
statistic, pvalue = stats.ranksums(df['A'], df['B'], alternative='two-sided')
print(round(statistic, 4), round(pvalue, 4))
# 6. 귀무가설 기각여부 결정
# p-value 가 0.05보다 크므로 귀무가설 채택
# 즉, A와 B의 혈압평균은 같다고 할 수 있다
# 답 : 채택
```

### 문제 7.
다음은 A그룹과 B그룹 인원의 혈압 데이터이다. A그룹의 혈압 평균이 B그룹보다 크다고 할 수 있는지
독립표본 t검정을 실시하시오. (유의수준 5%)

```python
#	데이터	만들기
df = pd.DataFrame({
				'A': [120,	135,	122,	124,	135,	122,	145,	160,	155,	142,	144,	135,	167],
				'B'	: [110,	132,	123,	119,	123,	115,	140,	162,	142,	138,	135,	142,	160]})
#print(df.head(3))
# 1. 가설설정
# H0 : A그룹의 혈압 평균이 B그룹보다 작거나 같다. A <= B
# H1 : A그룹의 혈압 평균이 B그룹보다 크다. A > B
# 2. 유의수준 확인 : 유의수준 5%로 확인
# 3. 정규성 검정
# H0 : 정규분포를 따른다
# H1 : 정규분포를 따르지 않는다
import scipy.stats as stats
from scipy.stats import shapiro
statisticA, pvalueA = stats.shapiro(df['A'])
statisticB, pvalueB = stats.shapiro(df['B'])
print(round(statisticA, 4), round(pvalueA, 4))
print(round(statisticB, 4), round(pvalueB, 4))
# 정규분포를 따른다
# 4. 등분산 검정
# H0 : 등분산 한다
# H1 : 등분산 하지 않는다
statistic, pvalue = stats.bartlett(df['A'], df['B'])
print(round(statistic, 4), round(pvalue, 4))
# 등분산 한다
# 5-1. 검정실시 (정규성O)
statistic, pvalue = stats.ttest_ind(df['A'], df['B'], equal_var=True, alternative='greater')
print(round(statistic, 4), round(pvalue, 4))
# 5-2. (정규성X)
statistic, pvalue = stats.ranksums(df['A'], df['B'], alternative='greater')
print(round(statistic, 4), round(pvalue, 4))
# 6. 귀무가설 기각여부 결정
# p-value가 0.05보다 크므로 귀무가설 채택
# 즉, A 혈압 평균이 B 혈압 평균보다 크다고 할 수 없다. 작거나 같다
# 답 : 채택
```

### 문제 8.
다음은 A, B, C 그룹 인원 성적 데이터이다.
세 그룹의 성적 평균이 같다고 할 수 있는지 ANOVA 분석을 실시하시오. (유의수준 5%)

```python
 #	데이터	만들기
df = pd.DataFrame(	{
	'A': [120,	135,	122,	124,	135,	122,	145,	160,	155,	142,	144,	135,	167],
	'B'	: [110,	132,	123,	119,	123,	115,	140,	162,	142,	138,	135,	142,	160],
	'C'	: [130,	120,	115,	122,	133,	144,	122,	120,	110,	134,	125,	122,	122]})
#print(df.head(3))
# 1. 가설설정
# H0 : 세 그룹의 성적 평균이 같다. A = B = C
# H1 : 적어도 하나는 같지 않다. not H0
# 2. 유의수준 확인 : 유의수준 5%로 확인
# 3. 정규성 검정
import scipy.stats as stats
from scipy.stats import shapiro
statisticA, pvalueA = stats.shapiro(df['A'])
statisticB, pvalueB = stats.shapiro(df['B'])
statisticC, pvalueC = stats.shapiro(df['C'])
print(round(statisticA, 4), round(pvalueA, 4))
print(round(statisticB, 4), round(pvalueB, 4))
print(round(statisticC, 4), round(pvalueC, 4))
# 정규분포를 따른다.
# 4. 등분산 검정
statistic, pvalue = stats.bartlett(df['A'], df['B'], df['C'])
print(round(statistic, 4), round(pvalue, 4))
# 등분산 한다.
# 5-1. 검정실시 (정규성O, 등분산O)
statistic, pvalue = stats.f_oneway(df['A'], df['B'], df['C'])
print(round(statistic, 4), round(pvalue, 4))
# (정규성O, 등분산X) welch_anova
# 5-2. (정규성X) 크루스칼 왈리스 검정
statistic, pvalue = stats.kruskal(df['A'], df['B'], df['C'])
print(round(statistic, 4), round(pvalue, 4))
# 6. 귀무가설 기각여부 결정
# p-value 가 0.05보다 작으므로 귀무가설 기각
# 즉, 적어도 하나는 같지 않다
# 답 : 기각
```

### 문제 9.
랜덤 박스에 상품 A,B,C,D 가 들어있다.
다음은 랜덤박스에서 100번 상품을 꺼냈을 때의 상품 데이터라고 할 때
상품이 동일한 비율로 들어있다고 할 수 있는지 검정해보시오.

```python
 #	데이터	생성
row1 = [30,	20,	15,	35]
df = pd.DataFrame([row1], columns=['A','B','C',	'D'])
#df
# 1. 가설설정
# H0 : 상품이 동일한 비율로 들어있다
# H1 : 상품이 동일한 비율로 들어있지 않다
# 2. 유의수준 확인 (유의수준 5%)
# 3. 검정실시
import scipy.stats as stats
from scipy.stats import chisquare
statistic, pvalue = stats.chisquare(f_obs=df.iloc[0], f_exp=[25,25,25,25])
print(round(statistic, 4), round(pvalue, 4))
# 4. 귀무가설 기각여부 결정
# p-value 가 0.05보다 작으므로 귀무가설 기각
# 즉, 동일한 비율로 들어있지 않다.
# 답 : 기각
```

### 문제 10.
랜덤 박스에 상품 A,B,C가 들어있다.
다음은 랜덤박스에서 150번 상품을 꺼냈을 때의 상품 데이터라고 할 때
상품별로 A 30%, B 15%, C 55% 비율로 들어있다고 할 수 있는지 검정해보시오. 

```python
 #	데이터	생성
row1 = [50,25,75]
df = pd.DataFrame([row1], columns=['A','B','C'])
#df
# 1. 가설설정
# H0 : 상품별로 30, 15, 55 비율로 들어있다
# H1 : 상품별로 30, 15, 55 비율로 들어있지 않다
# 2. 유의수준 확인 : 유의수준 5%로 확인
# 3. 검정실시
import scipy.stats as stats
from scipy.stats import chisquare
statistic, pvalue = chisquare(f_obs=df.iloc[0], f_exp=[150*0.3, 150*0.15, 150*0.55])
print(round(statistic, 4), round(pvalue, 4))
# 4. 귀무가설 기각여부 결정
# p-value 가 0.05보다 크므로 귀무가설 채택
# 즉, 30,15,55 비율로 들어있다고 할 수 있다
# 답 : 채택
```

- 2-1) 귀무가설과 대립가설을 설정하시오.
- 2-2) 상관계수 값을 구하시오.
- 2-3) 검정통계량과 p-value 값을 구하시오.
- 2-4) 귀무가설 채택/기각 여부를 결정하시오.

```python
###############		실기환경	복사	영역		###############
 #	데이터	불러오기
import	pandas	as	pd
import	numpy	as	np
#	실기	시험	데이터셋으로	셋팅하기	(수정금지)
from	sklearn.datasets	import	load_diabetes
 #	diabetes	데이터셋	로드
diabetes	=	load_diabetes()
x	=	pd.DataFrame(diabetes.data,	columns=diabetes.feature_names)
y	=	pd.DataFrame(diabetes.target)
y.columns	=	['target']
 ###############		실기환경	복사	영역		###############
x = x['bmi']
y = y['target']

# 1. 가설설정
# H0 : 두 변수는 선형관계가 없다. p = 0
# H1 : 두 변수는 선형관계가 있다. p != 0
# 2. 유의수준 확인 : 유의수준 5%로 확인
# 3. 검정실시
import scipy.stats as stats
from scipy.stats import pearsonr

r, pvalue = pearsonr(x, y)
print(r)
print(pvalue)
n = len(x)

statistic = r * np.sqrt((n-2)/(1-r**2))
print(statistic)
# 4. 귀무가설 기각여부 결정
# p-value 가 0.05보다 작으므로 귀무가설 기각
# 즉, 선형관계가 있다.
# 답 : 기각
```

```python
# 데이터 생성
import pandas as pd
import numpy as np
import seaborn as sns
df = sns.load_dataset('titanic')

df = df[['survived','sex','sibsp','fare']]
# sex:성별, sibsp:탑승한 부모 및 자녀 수, fare:요금
df['sex'] = df['sex'].map({'female':1, 'male':0})

x_train = df.drop(columns=['survived']).loc[0:499]
x_test = df.drop(columns=['survived']).loc[500:]
y_train = df['survived'].loc[0:499]
y_test = df['survived'].loc[500:]

# GLM 모델
import statsmodels.api as sm

x_train = sm.add_constant(x_train)
model = sm.GLM(y_train, x_train, family=sm.families.Binomial()).fit()
summary = model.summary()
print(summary)

x_test = sm.add_constant(x_test)

y_result_prob = model.predict(x_test)
#print(y_result_prob)
y_result = np.where(y_result_prob>0.5, 1, 0)
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_result)
print(acc)
print(round(1-acc, 2))
```
