```python
# 강의 풀이
#	hp	항목의	상위	7번째	값으로	상위	7개	값을	변환한	후,
#	hp가	150	이상인	데이터를	추출하여	hp의	평균값을	반올림하여	정수로	출력하시오.
df = df.sort_values('hp', ascending=False)
df = df.reset_index(drop=True)
top7 = df['hp'].loc[6]

import numpy as np
df['hp'] = np.where(df['hp']>top7, top7, df['hp'])
cond1 = (df['hp']>=150)
df = df[cond1]
print(round(df['hp'].mean()))
```
