```python
###############  실기환경 복사 영역  ###############
import pandas as pd
import numpy as np

# 실기 시험 데이터셋으로 셋팅하기 (수정금지)
from sklearn.datasets import load_wine
# 와인 데이터셋 load 
X, y = load_wine(return_X_y= True, as_frame=True)

# 실기 시험 데이터셋으로 셋팅하기 (수정금지)
from sklearn.model_selection import train_test_split
x_tr, x_te, y_tr, y_te = train_test_split(X, y, test_size=0.2,
                                          stratify=y,
                                          random_state=2024)
test = pd.DataFrame(x_te)
x_tr = pd.DataFrame(x_tr)
y_tr = pd.DataFrame(y_tr)
y_tr.columns = ['class'] 
y_test = pd.DataFrame(y_te)
train = pd.concat([y_tr, x_tr], axis=1)

train['hue'] = np.where(train['hue']>=1, 'H', 'L')
test['hue'] = np.where(test['hue']>=1, 'H', 'L')
train.iloc[0:3, train.columns.get_loc('alcohol')] = np.nan
test.iloc[0:3, test.columns.get_loc('alcohol')] = np.nan
###############  실기환경 복사 영역  ###############
# print(train)
# print(test)
x_train = train.drop(columns=['class'])
x_test = test
y_train = train['class']

# 1. 라이브러리 및 데이터 확인
# print(x_train.head(3))
# print(x_test.head(3))
# print(y_train.head(3))
# print(x_train.info())
# print(x_test.info())
# print(y_train.info())

# 2. 데이터 탐색(EDA)
# print(x_train.describe().T)
# print(x_test.describe().T)
# hue object
# print(x_train.describe(include='object').T)
# print(x_test.describe(include='object').T)
#print(y_train.value_counts())
# print(x_train.isnull().sum())
# print(x_test.isnull().sum())
# print(y_train.isnull().sum()) # train, test 모두 alcohol 결측치

# 3. 데이터 전처리 및 분리
# 1)결측치 2)이상치 3)변수 처리
med = x_train['alcohol'].median()
x_train['alcohol'] = x_train['alcohol'].fillna(med)
x_test['alcohol'] = x_test['alcohol'].fillna(med)
# print(x_train.isnull().sum())
# print(x_test.isnull().sum())
# print(y_train.isnull().sum()) # train, test 모두 alcohol 결측치
x_train = pd.get_dummies(x_train)
x_test = pd.get_dummies(x_test)
# print(x_train.info())
# print(x_test.info())
# 데이터 분리
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x_train,
                                                  y_train,
                                                  test_size=0.2,
                                                  stratify=y_train,
                                                  random_state=2024)
# print(x_train.shape)
# print(x_val.shape)
# print(y_train.shape)
# print(y_val.shape)
# x_test = x_test.reindex(columns=x_train.columns, fill_value=0)

# 4. 모델링 및 성능평가
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=2024)
model.fit(x_train, y_train)
y_pred = model.predict(x_val)

from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
acc = accuracy_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred, average='macro')
# print(acc)
# print(f1)

y_result = model.predict(x_test)
y_result_prob = model.predict_proba(x_test)
# print(y_result[:5])
# print(y_result_prob[:5])
result = pd.DataFrame({'result':y_result,
                       'prob0':y_result_prob[:,0],
                       'prob1':y_result_prob[:,1],
                       'prob2':y_result_prob[:,2]})
#print(result[:5])

# 5. 예측값 제출
pd.DataFrame({'pred':y_result}).to_csv('result_모의3.csv', index=False)

df = pd.read_csv('result_모의3.csv')
#df.head(3)


##### 실제 성능평가
from sklearn.metrics import accuracy_score, f1_score
acc2 = accuracy_score(y_test,y_result)
f12 = f1_score(y_test,y_result, average='macro')
print(acc2)
print(f12)
```
