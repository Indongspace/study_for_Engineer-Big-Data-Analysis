### 2) 열(Columns) 추가/제거

```python
# 열 선택
df_cyl = df['cyl']
df_cyl.head(3)
```

```python
df_new = df[['cyl','mpg']]
df_new.head(3)
```

```python
# 열 제거
df.drop(columns=['car','mpg','cyl']).head(3)
```

```python
# 열 추가
df2 = df.copy() # 이렇게 복사해야 함 (df2 = df 복사는 df가 바뀌면 df2도 바뀜)
df2['new'] = df2['mpg'] + 10
df2.head(3)
```

### 3) 데이터 필터링

```python
# 1개 조건 필터링
# cyl = 4 인 데이터의 수
cond1 = (df['cyl']==4)
len(df[cond1])
```

```python
len(df[df['cyl']==4])
```

```python
# mpg가 22 이상인 데이터 수
cond2 = (df['mpg']>=22)
print(len(df[cond2]))
```

```python
len(df[df['mpg']>=22])
```

```python
# 2개 조건 필터링
df[cond1&cond2]
```

```python
# 2개 조건 필터링 후 데이터 개수(and)
print(len(df[cond1&cond2]))
```

```python
# 2개 조건 필터링 후 데이터 개수(or)
df[cond1|cond2]
```

```python
print(len(df[cond1|cond2]))
```

```python
# 한번에 코딩할 경우
print(len(df[(df['cyl']==4) & (df['mpg']>=22)]))
print(len(df[(df['cyl']==4) | (df['mpg']>=22)]))
```

### 4) 데이터 정렬

```python
# 내림차순 정렬 (위에서부터 내려간다)
df.sort_values('mpg',ascending=False).head()
```

```python
# 오름차순 정렬 (밑에서부터 올라간다)
df.sort_values('mpg',ascending=True).head()
```

### 5) 데이터 변경(조건문)

```python
# np.where 활용
# hp 변수 값 중에서 205가 넘는 값은 205로 처리하고 나머지는 그대로 유지
import pandas as pd
import numpy as np
df = pd.read_csv('mtcars.csv')
df['hp'] = np.where(df['hp']>=205, 205, df['hp'])

# 내림차순 정렬
df.sort_values('hp', ascending=False).head()

# 활용 : 이상치를 Max 값이나 Min 값으로 대체할 경우 조건문 활용
```
