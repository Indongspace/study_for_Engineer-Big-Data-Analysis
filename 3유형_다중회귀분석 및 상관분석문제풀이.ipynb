### 당뇨병 환자의 질병 진행정도 데이터셋

```python
###############		실기환경	복사	영역		###############
 #	데이터	불러오기
import	pandas	as	pd
import	numpy	as	np
#	실기	시험	데이터셋으로	셋팅하기	(수정금지)
from	sklearn.datasets	import	load_diabetes
 #	diabetes	데이터셋	로드
diabetes	=	load_diabetes()
x	=	pd.DataFrame(diabetes.data,	columns=diabetes.feature_names)
y	=	pd.DataFrame(diabetes.target)
y.columns	=	['target']
 ###############		실기환경	복사	영역		###############
```

```python
print(diabetes.DESCR)
```

## 1. sklearn 라이브러리 활용

```python
from sklearn.linear_model import LinearRegression
```

```python
# 독립변수와 종속변수 설정
x = x[['age','sex','bmi']]
print(x.head())
print(y.head())
```

- 회귀식 : y = b0 + b1x1 + b2x2 + b3x3
- (x1 = age, x2 = sex, x3 = bmi)

```python
# 모델링
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
# 회귀분석 관련 지표 출력

# 1. Rsq(결정계수) : model.score(x, y)
model.score(x,y)
print(round(model.score(x,y), 2))
```

```python
# 2. 회귀계수 출력 : model.coef_
print(np.round(model.coef_, 2)) # 전체 회귀계수
print(np.round(model.coef_[0,0], 2)) # x1의 회귀계수
print(np.round(model.coef_[0,1], 2)) # x2의 회귀계수
print(np.round(model.coef_[0,2], 2)) # x3의 회귀계수
```

```python
# 3. 회귀계수(절편) : model.intercept_
print(np.round(model.intercept_, 2))
```

## 상관분석

```python
###############		실기환경	복사	영역		###############
 #	데이터	불러오기
import	pandas	as	pd
import	numpy	as	np
#	실기	시험	데이터셋으로	셋팅하기	(수정금지)
from	sklearn.datasets	import	load_diabetes
 #	diabetes	데이터셋	로드
diabetes	=	load_diabetes()
x	=	pd.DataFrame(diabetes.data,	columns=diabetes.feature_names)
y	=	pd.DataFrame(diabetes.target)
y.columns	=	['target']
 ###############		실기환경	복사	영역		###############
```

```python
# 상관분석을 할 2가지 변수 설정
x = x['bmi']
y = y['target']
print(x.head())
print(y.head())
```

```python
# 라이브러리 불러오기
from scipy.stats import pearsonr

# 상관계수에 대한 검정실시
r, pvalue = pearsonr(x,y)

# 가설설정
# H0 : 두 변수간 선형관계가 존재하지 않는다. (p = 0)
# H1 : 두 변수간 선형관계가 존재한다. (p != 0)

# 1. 상관계수
print(round(r, 2))
# 2. p-value
print(round(pvalue, 2))

# 3. 검정통계량
# 통계량은 별도로 구해야 함 (T = r * root(n-2) / root(1-r^2))
# r = 상관계수
# n = 데이터의 개수
n = len(x) # 데이터 수
r2 = r**2 # 상관계수의 제곱
statistic = r*((n-2)**0.5) / ((1-r2)**0.5)

print(round(statistic, 2))
```

```python
# 4. 귀무가설 기각여부 결정(채택/기각)
# p-value 값이 0.05보다 작기 때문에 귀무가설을 기각한다.(대립가설채택)
# 즉, 두 변수간 선형관계가 존재한다고 할 수 있다.(상관계수가 0이 아니다)

# 답 : 기각
```
